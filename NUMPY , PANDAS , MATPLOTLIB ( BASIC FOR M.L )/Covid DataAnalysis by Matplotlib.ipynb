```python
import pandas as pd
import urllib.request as url
import matplotlib.pyplot as plt
import numpy as np
import json
plt.style.use('ggplot')
```

```python
res = url.urlopen('https://api.covid19india.org/states_daily.json')
```

```python
data = json.load(res)
```

```python
df = pd.DataFrame(data['states_daily'])
```

```python
df['dateymd'][0]   #it will string date as a string
                  #it will works as a single string
```

```python
df['tt'][0]
```

```python
 df['dateymd'] = pd.to_datetime(df['dateymd'])
```

```python
df['dateymd'][0]  #coverted into date time format
```

```python
df['dl'] = df['dl'].astype('int32') #it will return a integer value
```

```python
df['tt'] = df['tt'].astype('int32')
df['mh'] = df['mh'].astype('int32')
df['rj'] = df['rj'].astype('int32')
```

```python
confirmed_df= df[df['status'] == 'Confirmed']
recovered_df= df[df['status'] == 'Recovered']
deceased_df= df[df['status'] == 'Deceased']
```
