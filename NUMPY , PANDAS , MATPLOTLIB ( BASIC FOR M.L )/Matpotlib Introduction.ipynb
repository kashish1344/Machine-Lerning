```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
     'names' : ['Sachin','Virat','Dhoni','Yuvraj','Rohit','Dhawan'],
     'score' : [123,67,29,78,63,19],
     'balls' : [100,89,56,45,87,56]
     
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10,6))   #(Width,Height)
plt.title('Scores By Each Player')
plt.bar(df['names'],df['score'],width=0.9,color='r')
plt.xlabel("Name")
plt.ylabel('Score')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))   #(Width,Height)
plt.title('Scores By Each Player')
plt.bar(df['names'],df['score'],width=0.9,color='r')
plt.xlabel("Name")
plt.ylabel('Score')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))   #(Width,Height)
plt.title('Balls Taken By Each Player')
plt.barh(df['names'],df['balls'])
plt.xlabel("Name")
plt.ylabel('Balls')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Balls Taken By Each Player')
plt.hist(df['score'],bins=[i for i in range(1,100,20)],rwidth=0.9 ,color='r')
plt.xlabel("Scoes")
plt.ylabel('Frequency')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')
plt.plot(df['balls'],df['score'])
plt.xlabel("Balls")
plt.ylabel('Scores')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')    # Scatter Ploting
plt.plot(df['balls'],df['score'],'o')
plt.xlabel("Balls")
plt.ylabel('Scores')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')    # Scatter Ploting
plt.plot(df['balls'],df['score'],'bo')   # bo means blue Scatter
plt.xlabel("Balls")
plt.ylabel('Scores')
plt.show()
```

```python
 plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')    # Using  Scatter Ploting
plt.scatter(df['balls'],df['score']) 
plt.xlabel("Balls")
plt.ylabel('Scores')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')   
plt.pie(x=df['score'],labels=df['names']) 
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Balls Taken By Each Player')   
plt.pie(x=df['balls'],labels=df['names'], shadow=True) 
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')   
plt.pie(x=df['score'],labels=df['names'], shadow=True, 
        autopct='%1.2f%%',explode=(0,0,0,0.3,0,0)) 
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')   
plt.pie(x=df['score'],labels=df['names'], shadow=True, 
        autopct='%1.2f%%',explode=(0,0,0,0.3,0,0),startangle=90) 
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')   
plt.pie(x=df['score'], labels=df['names'], shadow=True, 
        autopct='%1.2f%%', explode=(0,0,0,0.3,0,0),
        startangle=90, counterclock=False) 
plt.show()    #counter Clock is for starting from the first Name
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))  
plt.title('Scores By Each Player')    
plt.boxplot(df['score']) 
plt.xlabel("Balls")
plt.ylabel('Scores')
plt.show()
```
