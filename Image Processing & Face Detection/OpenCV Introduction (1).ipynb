```python
# Blur Image
#types:-
# - Gausian Blur
# -  Median Blur
# - Bilateral Blur

img = cv2.imread('Dhoni.jpg')
img = cv2.resize(img,None,fx=0.5, fy=0.5)

img = cv2.GaussianBlur(img,(15,15),100)

cv2.imshow('result',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
